```python
from itertools import cycle
switch = cycle(('x','o'))
print("Welcome to Tic Tac Toe App!\n\n")
input1=['_','_','_','_','_','_','_','_','_']
def display(input1):
    if len(input1)==9:
        print(f"{input1[0]} || {input1[1]} || {input1[2]}")
        print('-----------')
        print(f"{input1[3]} || {input1[4]} || {input1[5]}")
        print('-----------')
        print(f"{input1[6]} || {input1[7]} || {input1[8]}")
position=list(range(1,10,1))
display(position)
n=[]
def x_turn():
    try:
        x_place=int(input("X:Where do you want to place the piece? (1-9) "))
        if x_place>0 and x_place<10:
            if x_place not in n:
                n.append(x_place)
                input1[x_place-1]='X'  
                display(input1)
            else:
                print("Don't enter a repeating number.")
        else:
            print("Enter a number between 1 and 9...")
    except ValueError:
        print("Enter a number...")
def o_turn():
    try:
        o_place=int(input("O:Where do you want to place the piece? (1-9) "))
        if o_place>0 and o_place<10:
            if o_place not in n:
                n.append(o_place)
                input1[o_place-1]='O'  
                display(input1)
            else:
                print("Don't enter a repeating number.")
        else:
            print("Enter a number between 1 and 9...")
    except ValueError:
        print("Enter a number...")
def win():
    for i in range(3):
        if input1[3*i]==input1[3*i+1] and input1[3*i+2]==input1[3*i]:
            if input1[3*i]!='_':
                print(input1[3*i].upper() + " Wins!!")
                return True
        elif input1[i]==input1[i+3] and input1[i+6]==input1[i]:
            if input1[i]!='_':
                print(input1[i].upper() + " Wins")
                return True
        elif input1[0]==input1[4] and input1[8]==input1[0]:
            if input1[4]!='_':
                print(input1[4].upper() + " Wins!!!")
                return True
        elif input1[2]==input1[4] and input1[6]==input1[2]:
            if input1[2]!='_':
                print(input1[4].upper() + " Wins!!!")
                return True
i=0
while True:
    choice = next(switch)
    if choice=='x':
        x_turn()
    elif choice=='o':
        o_turn()
    winner=win()
    if winner:
        break
    elif '_' not in input1:
        print("It's a Tie...")
        break
    i+=1
```
